```python
# to get consumer_key, consumer_secret, access_token and access_secret from twitter for WeRateDogs dataset
consumer_key = 'CONSUMER KEY'
consumer_secret = 'CONSUMER SECRET'
access_token = 'ACCESS TOKEN'
access_secret = 'ACCESS SECRET'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

# Construct the API instance firstly to parse the result to json object, secondly to wait for rate limits and lastly to print
#a notification when tweepy is waiting for rate limits
api = tweepy.API(auth, 
                 parser = tweepy.parsers.JSONParser(), 
                 wait_on_rate_limit = True, 
                 wait_on_rate_limit_notify = True) 
```

```python
tweet_json=open('tweet_json.txt','w')
#txt_file.write('tweet_id retweet_count favorite_count''\n')
for tweet_id in df.tweet_id:
    try:
        tweet_info=api.get_status(str(tweet_id),tweet_mode='extended')
        retweet_count=tweet_info.retweet_count
        print(tweet_id)
        favorite_count=tweet_info.favorite_count
        tweet_json.write(tweet_id+' '+str(retweet_count)+' '+str(favorite_count)+'\n')  
    except Exception as e:
        print('error')
tweet_json.close()
```

```python
df_tweet_json = pd.DataFrame(columns=['tweet_id','retweet_count','favorite_count'])
with open('tweet_json.txt','r')as file:
    for line in file:
        tweet_id,retweet_count,favorite_count=line[:-1].split(' ')
        df_tweet_json = df_tweet_json.append({'tweet_id':tweet_id,
                                              'retweet_count':retweet_count,
                                              'favorite_count':favorite_count},
                                               ignore_index=True)
```

## Define

### To join all the 3 datasets

## Code

```python
df_clean.tweet_id = df_clean.tweet_id.astype(str)
df_images_clean.tweet_id = df_images_clean.tweet_id.astype(str)
df_tweet_json_clean.tweet_id = df_tweet_json_clean.tweet_id.astype(str)

df_clean = pd.merge(df_clean, df_images_clean,
                        how = 'inner', on = 'tweet_id')
```

```python
df_tweet_json_clean = df_tweet_json_clean[~(df_tweet_json_clean.tweet_id.duplicated())]
df_clean=pd.merge(df_clean,df_tweet_json_clean,on='tweet_id',how='left')
df_clean[df_clean.tweet_id=='887473957103951883']
missingcount_id=df_clean.tweet_id[df_clean.retweet_count.isnull()]
```

## Define

### Combine all the dog columns and put them in one column

## Code

```python
#firstly replace all the data from the doggo, floofer, pupper and puppo columns and replace them wth none
df_clean.doggo=df_clean.doggo.replace('None','')
df_clean.floofer=df_clean.floofer.replace('None','')
df_clean.pupper=df_clean.pupper.replace('None','')
df_clean.puppo=df_clean.puppo.replace('None','')
```

```python
#then create a new column named dog_category and put them into the new column
df_clean['dog_category']=df_clean.doggo.str.cat(df_clean.floofer).str.cat(df_clean.pupper).str.cat(df_clean.puppo)
```

```python
#then we drop all four previous columns as ther are of no use now
df_clean=df_clean.drop(['doggo','floofer','pupper','puppo'],axis=1)
df_clean.dog_category=df_clean.dog_category.replace('',np.nan)
```
